```python
import os
import tensorflow as tf
from model import *
from data import *
from dataloader import DataLoader

os.environ["CUDA_VISIBLE_DEVICES"]="1"
```

```python
taxa = 'spider' # choose a taxa
train_dataloader_args = dict(
    input_height = 560,
    input_width = 336,
    data_folder = './data/data',
    samples_json = f'samples_{taxa}.json',
    env_json = 'env.json',
    bio_json = 'bio_taxa.json',
    batch_size = 4,
    input_step_size=4,
    output_step_size=4,
    shuffle=True,
    aug=False,
    train=True
)
```

```python
train_dataloader = DataLoader(train_dataloader_args)

input_channel = train_dataloader.image_input_channel
output_channel = train_dataloader.output_channel
model = unet(
    image_input_size=(train_dataloader_args['input_height'], train_dataloader_args['input_width'], input_channel), 
    vector_input_size=(1, 1, train_dataloader.vector_input_channel),
    output_size=(train_dataloader_args['input_height'], train_dataloader_args['input_width'], output_channel),
)
model_checkpoint = ModelCheckpoint(
    f'unet_{taxa}.hdf5', 
    monitor='loss',
    verbose=1, 
    save_best_only=False,
    save_freq='epoch'
)
```

```python
model.fit(
    train_dataloader,
    epochs=60,
    callbacks=[model_checkpoint]
)
```

```python
tf.keras.utils.plot_model(
    model,
    to_file="model.png",
    show_shapes=False,
    show_dtype=False,
    show_layer_names=True,
    rankdir="TB",
    expand_nested=False,
    dpi=128,
    layer_range=None,
)
```
